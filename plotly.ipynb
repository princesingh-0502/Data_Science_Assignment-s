```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import cufflinks as cf
cf.go_offline(connected=True)

import warnings
warnings.filterwarnings('ignore')
```

```python
# datasets 

iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
mpg = sns.load_dataset('mpg')
gap = px.data.gapminder()
attention = sns.load_dataset('attention')
flights = sns.load_dataset('flights')
wind = px.data.wind()
stocks = px.data.stocks()
tips = sns.load_dataset('tips')
```

```python
df = pd.DataFrame(np.random.randn(50,4),columns=['A','B','C','D'])
plt.figure(figsize=(7,1))
df.plot()
plt.show()
```

```python
# df.iplot()
px.line(df)
```

```python
# timeseries 
# bi variate analysis

stocks
stock_colours=['#A2AAAD','#00A4EF','#E50914','#4285F4']
px.line(stocks,x='date',y=['AAPL','MSFT','NFLX','GOOG'],
        height=400,title='stock prices(standardised)ovewr the year',
        labels={'value':'stock prices','date':'Date'},
        color_discrete_sequence=stock_colours)
```

```python
padosi = gap[gap.country.isin(['India','Bangladesh','China','Nepal','Sri Lanka'])]
padosi
palette = ['#006A4E','#DE2910','#FF9933','#DC143C','#FFB700']
px.line(padosi,x='year',y='lifeExp',
        color='country',color_discrete_sequence=palette,
        hover_data=gap.columns,width=800,height=500,
        title='life expentancy over the year')
```

```python
tips
sex_palette = ['#E64980','#4C6EF5']
px.bar(tips,x='day',y='total_bill',color='sex',
        color_discrete_sequence=sex_palette,
        barmode='group',width=800,height=400)
```

```python
import plotly.colors
europe=gap.query('continent=="Europe"')
# animination plot 
px.bar(europe,x='country',y='pop',animation_frame='year',color='country',
      text='pop',color_discrete_sequence=plotly.colors.sequential.RdBu,height=600).update_traces(texttemplate='%{text:.2s}',textposition='outside')
```

```python
iris
px.scatter(iris,x='petal_length',y='petal_width',color='species',width=800)
```

```python
import plotly.express as px

fig = px.scatter(tips,x='total_bill',y='tip',color='sex',size='size',hover_data=tips.columns)
fig.show()
```

```python
# pie chart 
# bi variate analysis
# unvariate analysis
temp_who = titanic.who.value_counts().reset_index()

px.pie(temp_who,values='count',names='who',
      color_discrete_sequence=plotly.colors.sequential.RdBu)
```

```python
asia_pop = gap.query('continent=="Asia"').query('year==2007').sort_values(by='pop',ascending=False)

asia_pop.head(10)

px.pie(asia_pop.head(10),values='pop',names='country',
      color_discrete_sequence=plotly.colors.sequential.RdBu)
```
