# 4. Explain which method (mean or median) is more appropriate and why

In most cases, the median is more appropriate than the mean for filling missing values in the Age column. This is because age data in datasets like the Titanic dataset is typically skewed, meaning the values are not evenly distributed and may include extreme ages (very young or very old passengers). The mean is sensitive to such outliers, so a few unusually high or low values can shift the average and lead to less accurate imputation. In contrast, the median represents the middle value of the distribution and is not affected by extreme values, making it more robust and reliable for skewed data. Therefore, median imputation preserves the true central tendency of the Age variable more effectively and is generally the better choice

# 7. Explain why skewed distributions affect mean.

#Answer
The mean is calculated by adding all values and dividing by the total number of observations. Because it uses every value, it is highly sensitive to extreme values (outliers).

🔹 In a Right-Skewed Distribution (Positive Skew)

There are a few very large values.

These large values pull the mean toward the right.

Result:
Mean > Median

Example (like Titanic Fare):
A few passengers paid very high fares → Mean increases significantly.

🔹 In a Left-Skewed Distribution (Negative Skew)

There are a few very small values.

These small values pull the mean toward the left.

Result:
Mean < Median

🔹 Why This Happens

The mean is calculated as:

Mean
=
Sum of all values
Total number of values
Mean=
Total number of values
Sum of all values
	


Even a few extreme values can significantly change the total sum, which shifts the mean.

# **SECTION 8: Conceptual Questions**

# 1. What is the difference between mean and median imputation?

**Mean Imputation:** Uses the average. It is best for symmetric, normally distributed, continuous data. It preserves the mean of the dataset but can be heavily influenced by outliers, which can bias the imputation.

**Median Imputation:** Uses the middle value. It is best for skewed numerical data or datasets with outliers. It is more robust than mean imputation, meaning the imputed values are less affected by extreme, non-representative data points

# 3. Why is median preferred in skewed distributions?

🔹 1. Median is Not Affected by Outliers

The median is the middle value when data is arranged in order.
It does not depend on extreme values.

Example:
Data = 10, 12, 15, 18, 200

Mean = 51 (affected by 200)

Median = 15 (represents actual center)

Here, 200 pulls the mean upward, but the median remains stable.

🔹 2. Mean Gets Shifted in Skewed Data

In right-skewed data, large values pull the mean to the right.

In left-skewed data, small values pull the mean to the left.

So the mean no longer represents the true central tendency.

🔹 3. Gives Better Central Representation

Since the median is based on position rather than magnitude, it gives a more accurate center in skewed datasets.

🔹 4. Inflated Dataset Size

🔹 5. Misleading Insights

# 5. What is the difference between loc and iloc?

The primary difference between loc and iloc in Pandas is that loc is label-based (using row/column names), while iloc is integer-position based (using 0-based numerical indices). loc includes the stop value in slices, whereas iloc excludes it. loc supports boolean arrays; iloc does not. 

Key Differences Summary:
Indexing Type: loc uses labels (e.g.'a''2023'), while iloc uses integer positions (e.g., 0, 1).
Slicing Behavior: df.loc[0:3] includes index 3, but df.iloc[0:3] excludes index 3 (returns positions 0, 1, 2).
Usage: Use loc for named rows/columns or boolean filtering. Use iloc for numerical indexing.
Data Types: loc accepts strings, integers (if they are labels), or boolean arrays. iloc accepts integers, lists of integers, or slice objects.
